# üìò Step 1: Setup & Imports

```python
# Install YOLOv8 (if not already)
# !pip install ultralytics --upgrade

from ultralytics import YOLO
import os
import matplotlib.pyplot as plt
import yaml
import cv2
```


# üìò Step 2: Configuration Block

```python

# Define dataset and training paths
dataset_yaml = "/home/ai_train/skyhub_project/july/back_seat_model_training/yolo_seat_back_data_code/yolov8_split/dataset.yaml"
model_type = "yolov8n-seg.pt"  # or yolov8s-seg.pt, etc.
imgsz = 640
epochs = 100
batch = 16
project = "seat_segmentation"
name = "yolov8n_seg_run"
```


# üìò Step 3: Load Dataset Info & Summary

```python

with open(dataset_yaml, 'r') as f:
    data_yaml = yaml.safe_load(f)

class_names = data_yaml['names']
print(f"\nüßæ Classes in dataset ({len(class_names)}):")
for k, v in class_names.items():
    print(f"  {k}: {v}")

# Count number of images
train_images = os.listdir(os.path.join(data_yaml['path'], data_yaml['train']))
val_images = os.listdir(os.path.join(data_yaml['path'], data_yaml['val']))
print(f"\nüñºÔ∏è Training images: {len(train_images)}")
print(f"üñºÔ∏è Validation images: {len(val_images)}")
```

# üìò Step 4: Train YOLOv8-seg Model

```python

model = YOLO(model_type)  # Load pretrained YOLOv8 segmentation model

results = model.train(
    data=dataset_yaml,
    imgsz=imgsz,
    epochs=epochs,
    batch=batch,
    name=name,
    project=project,
    task="segment"
)
```

# üìò Step 5: Visualize Training Metrics

```python

metrics = results.results_dict
if metrics:
    print("\nüìä Final Training Results:")
    for k, v in metrics.items():
        print(f"{k}: {v:.4f}")
```


# üìò Step 6: Inference Test on Validation Samples

```python
from glob import glob
import random

val_dir = os.path.join(data_yaml['path'], data_yaml['val'])
val_sample = random.choice(glob(os.path.join(val_dir, '*.jpg')))
print("\nüîç Running inference on:", val_sample)

inference_results = model.predict(source=val_sample, save=True, save_txt=True)
```

# üìò Step 7: Save Best Model Path

```python

best_model_path = model.best
print("\n‚úÖ Best model saved at:", best_model_path)

# You can now use this model for testing, validation or export.
```
